## Updating the dataset
#### Now we will transform the dataset to make it more appropriate for our analysis because some features are not necessary.

```python
#For example, the 'promoted' columns take values 1 if promoted or 0 if not, but we can clearly see that none of our rows
#are affected because none of the posts are promoted.
#So, we can delete the features
df[df['promoted']==1]
```

```python
df=df.drop(['promoted'],axis=1)
```

```python
#We can do the same for the features 'hrs' because almost every rows (except for 778) are set on '24'
df[df['hrs']!=24].hrs.count()
```

```python
#Just to be sure we can quicly compute the corelation between 'hrs' and the 'output'.
#Correlation is almost 0, we can delete the features 'hrs'
df['hrs'].corr(df['output'])
```

```python
df=df.drop(['hrs'],axis=1)
```

```python
#Then, for the 'baseTime' feature, I didn't really understand the aim of it. It is range for values from 0 to 71. 
#According to the website where we are downloading the dataset, it is says that it is 'Selected time in order to simulate the scenario'

#I tried to contact the author of the dataset for a more precise description but he didn't respond.
#Finally, 'baseTime' and 'output' are only -0.22 correlated.
df['baseTime'].corr(df['output'])
#So, I have decided to delete this features because I don't understand it, and it is not the most important one.

df=df.drop(['baseTime'],axis=1)
```

# Shape data for ML

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X_train.shape
```

# Model of prediction

## SVR

```python
from sklearn import linear_model
from sklearn.svm import SVR


svr = SVR()
svr.fit(X_train, y_train)
```

```python
pred_svr=svr.predict(X_test)
```

```python
pred_svr=pred_svr.round()
```

```python
mse_svr=mean_squared_error(y_test, pred_svr)
print(np.sqrt(mse_svr))
```

```python
df.describe()
```

## Ridge


```python
from sklearn.linear_model import Ridge

ridge=Ridge()
ridge.fit(X_train,y_train)
```

```python
ridge_pred=ridge.predict(X_test)
ridge_pred=ridge_pred.round()
```

```python
mse_ridge=mean_squared_error(y_test, ridge_pred)
print(np.sqrt(mse_ridge))
```

## Elastic Net


```python
from sklearn.linear_model import ElasticNet

elastic=ElasticNet()
elastic.fit(X_train,y_train)
```

```python
elastic_pred=elastic.predict(X_test)
elastic_pred=elastic_pred.round()
```

```python
mse_elastic=mean_squared_error(y_test, elastic_pred)
print(np.sqrt(mse_elastic))
```

## Random Forest

```python
from sklearn import tree

random_forest=tree.DecisionTreeRegressor()
random_forest.fit(X_train,y_train)
```

```python
tree_pred=random_forest.predict(X_test)
tree_pred=np.round(tree_pred)
```

```python
mse_tree=mean_squared_error(y_test, tree_pred)
print(np.sqrt(mse_tree))
```

# Test all models with Grid Search

```python
ensemble_clf=[svr, ridge, elastic, random_forest]
```

```python
#'C':np.linspace(0.1,5,20)

param_svr={'degree':[2,3,4,5]}
param_ridge={'alpha':np.linspace(0,10,20),'max_iter':np.linspace(1000,5000,10), 'tol':[1], 'solver':['svd','lsqr'], 'normalize':[True, False]}
param_elastic={'alpha':np.linspace(0,5,20),'max_iter':np.linspace(1000,5000,10), 'tol':[1], 'normalize':[True, False]}
param_random_forest={'max_depth':np.linspace(1000,5000,10), 'min_samples_split':[2,3,4,5]}

parameters_list=[param_svr, param_ridge, param_elastic, param_random_forest]
```

```python
model_log=["_svr", "_ridge", "_elastic", "_random_forest"]
```

```python
best_score=0
best_params=None
model=None

def GridSearch():
    
    for i in range(len(ensemble_clf)):
        Grid=GridSearchCV(estimator=ensemble_clf[i], param_grid=parameters_list[i], 
                        cv=5, scoring='neg_mean_squared_error', verbose=3, n_jobs=-1).fit(X_train, y_train)

        print("Model : ",ensemble_clf[i])
        print("Best score : ",Grid.best_score_)
        print("Best params : ",Grid.best_params_)
    
GridSearch()
```

# TensorFlow

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.layers import Input, Dense, Activation,Dropout
from tensorflow.keras.models import Model

print(tf.__version__)
```

```python
input_layer = Input(shape=(X.shape[1],))
dense_layer_1 = Dense(100, activation='relu')(input_layer)
dense_layer_2 = Dense(50, activation='relu')(dense_layer_1)
dense_layer_3 = Dense(25, activation='relu')(dense_layer_2)
output = Dense(1)(dense_layer_3)

model = Model(inputs=input_layer, outputs=output)
model.compile(loss="mean_squared_error" , optimizer="adam", metrics=["mean_squared_error"])
```

```python
history = model.fit(X_train, y_train, batch_size=3, epochs=40, verbose=3, validation_split=0.25)
```

```python
pred = model.predict(X_test)
print(mean_squared_error(y_test,pred))
```
